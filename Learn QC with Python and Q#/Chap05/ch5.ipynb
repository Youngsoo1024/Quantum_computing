Chapter 5. Teleportation and entanglement: Moving quantum data around.

```python
from qutip.core.gates import hadamard_transform
import qutip
```

```python
ket_0 = qutip.basis(2,0)
ket_plus = hadamard_transform()*ket_0
initial_state = qutip.tensor(ket_plus, ket_0)
initial_state
```

```python
swap_matrix =qutip.swap(N=2,M=2)
swap_matrix*initial_state
```
