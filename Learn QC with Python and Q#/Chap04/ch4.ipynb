Chapter 4. Nonlocal Games: Working with multiple qubits

```python
import numpy as np 
ket0 = np.array([[1], [0]])
ket1 = np.array([[0], [1]]) 
np.kron(ket0, ket1) 
```

```python
import numpy as np 
ket0 = np.array([[1], [0]]) 
from functools import reduce 
reduce(np.kron, [ket0] * 4) 
```

```python
import numpy as np 
A = np.array([[1, 3], [5, 7]])
B = np.array([[2, 4], [6, 8]]) 
np.kron(A, B) 
```
